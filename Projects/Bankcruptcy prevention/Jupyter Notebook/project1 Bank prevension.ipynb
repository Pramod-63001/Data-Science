# split the X and Y variables:


```python
X = df.drop("class" , axis = 1)
y = df["class"]
```

```python
from sklearn.metrics import accuracy_score , f1_score , confusion_matrix , ConfusionMatrixDisplay
```

```python
model_dict = {"Model Name" :[] , "Training Accuracy" : [] , "Testing Accuracy" : [] , "F1 Score Training" : [] ,"F1 Score Testing" : [] ,  "Recall score Training" : [] ,"Recall score Testing" : [] , "Precision Score Training" :[] , "Precision Score Testing" :[]} 
```

# spliting The Data:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
def error_matrix(X_train , X_test , y_train , y_test , model , model_name) :
    from sklearn.metrics import accuracy_score , f1_score , recall_score , precision_score
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)

    accuarcy_training = round(accuracy_score(y_train , y_train_pred),2)
    accuracy_testing = round(accuracy_score(y_test , y_test_pred),2)

    f1_score_training =  round(f1_score(y_train , y_train_pred),2)
    f1_score_testing = round(f1_score(y_test , y_test_pred),2)

    training_recall = round(recall_score(y_train , y_train_pred),2)
    testing_recall = round(recall_score(y_test , y_test_pred),2)

    precision_training = round(precision_score(y_train , y_train_pred),2)
    precision_testing = round(precision_score(y_test , y_test_pred),2)

    
    
    print(f"Training Accuracy : {accuarcy_training}")
    print(f"Testing Accuracy : {accuracy_testing}")
    
    print(f"Training f1_score : {f1_score_training}")
    print(f"Testing f1_score : {f1_score_testing}")
    
    print(f"Training recall_score : {training_recall}")
    print(f"Testing recall_score : {testing_recall}")
    
    print(f"Training precision_score : {precision_training}")
    print(f"Testing precision_score : {precision_testing}")
    
    model_dict["Model Name"].append(model_name)
    model_dict["Training Accuracy"].append(accuarcy_training)
    model_dict["Testing Accuracy"].append(accuracy_testing)
    model_dict["F1 Score Training"].append(f1_score_training)
    model_dict["F1 Score Testing"].append(f1_score_testing)
    model_dict["Recall score Training"].append(training_recall)
    model_dict["Recall score Testing"].append(testing_recall)
    model_dict["Precision Score Training"].append(precision_training)
    model_dict["Precision Score Testing"].append(precision_testing)

```

```python
def train_model(X_train , X_test , y_train , y_test , model) : 
    model.fit(X_train , y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    return model , train_pred , test_pred 
```

#  LogisticRegression Model:

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR , train_pred , test_pred = train_model(X_train, X_test, y_train, y_test , LogisticRegression())
```

```python
print("Confusion Metrix for Training Data for LR")
cm = confusion_matrix(y_train, train_pred, labels=LR.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=LR.classes_)
disp.plot()
plt.show()
```

```python
print("Confusion Metrix for Testing Data for LR")
cm = confusion_matrix(y_test, test_pred, labels=LR.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=LR.classes_)
disp.plot()
plt.show()
```

```python
error_matrix(X_train , X_test , y_train , y_test , LR , "Logistic Regression")
```
