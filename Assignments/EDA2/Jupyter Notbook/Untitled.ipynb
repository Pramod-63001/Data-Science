```python
# Feature Engineering
```

```python
def qualification(value) :
    if(value in [" Masters" ,  " Doctorate"]) :
        return "Higher Study"
    elif(value in [' Some-college', ' Bachelors', ' Assoc-voc', ' Assoc-acdm']) :
        return "College Degree"
    else :
        return "School"
df["education"].apply(qualification).value_counts()
df_cat["qualification"] = df["education"].apply(qualification)
# pd.DataFrame(df["education"].value_counts()).index
```

```python
def US_Citizen(value) :
    return value == " United-States"
df["native_country"].apply(US_Citizen).value_counts()
df_cat["US_Citizen"] = df["native_country"].apply(US_Citizen)
```

```python
def Private_job(value) :
    return value == " Private"
df["workclass"].apply(Private_job).value_counts()
df_cat["Private Job"] = df["workclass"].apply(Private_job)
```

```python
def race_white(value) :
    return value == " White"
df["race"].apply(race_white).value_counts()
df_cat["race_white"] = df["race"].apply(race_white)
```

```python

df_cat = df_cat.drop(["workclass" , "education" , "native_country" , "race"] , axis =1)
```

```python
from sklearn.preprocessing import OrdinalEncoder
categories = []
my_cats = ["marital_status","occupation","relationship","race_white"]
for cols in my_cats :
    categories.append(list(pd.DataFrame(df_cat[cols].value_counts()).index)[::-1])
ORED = OrdinalEncoder(categories=categories)
df_cat[my_cats] = pd.DataFrame(ORED.fit_transform(df_cat[my_cats]) , columns=my_cats)
```

```python
from sklearn.preprocessing import OneHotEncoder
my_cats = ["sex","qualification"]
OHE = OneHotEncoder(drop="first")
df_cat[["sex","qualification1" , "qualification2" ]] = OHE.fit_transform(df_cat[my_cats]).toarray()
```
