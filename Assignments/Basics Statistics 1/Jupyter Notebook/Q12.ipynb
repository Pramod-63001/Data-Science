```python
import numpy as np
import pandas as pd
```

```python
my_list=[34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56]
```

```python
my_series = pd.Series(my_list)
```

```python
print("mean......>",my_series.mean())
```

```python
print("median......>",my_series.median())
```

```python
print("mode is......>",my_series.mode())
```

```python
print("skewness is....>",my_series.skew())
```

```python
print("kurtosis is ....>",my_series.kurt())
```

```python
print("std id ....>",my_series.std())
```

```python
print("variance.....>",my_series.var())
```
