```python
# Outlier detection
```

```python
Q1 = np.percentile(my_series , 25)
Q3 = np.percentile(my_series , 75)
```

```python
IQR = Q3 - Q1
```

```python
LW = Q1 - 1.5*IQR
UW = Q3 + 1.5*IQR
```

```python
# my_series[(my_series < LW) |  (my_series > UW)]
```

```python
for i in my_series :
    if(i < LW or i > UW) :
        print(i)
```
