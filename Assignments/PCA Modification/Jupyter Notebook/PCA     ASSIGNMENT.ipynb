```python
from sklearn.preprocessing import StandardScaler
```

```python
df.describe()
```

```python
scaler=StandardScaler()
```

```python
df_new = pd.DataFrame(data = scaler.fit_transform(df) , columns=df.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 5)
pca_values = pca.fit_transform(df_new)
pca_values.shape
```

```python
#loadings or weights
pca.components_
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
pca_df = pd.DataFrame(data = pca_values , columns=[f"PCA_{i}" for i in range(5)])
```

```python
pca_df
```

```python
from sklearn.cluster import KMeans
```

```python
KM = KMeans()
```

```python
KM.fit(df_new)
```

```python
pd.Series(KM.predict(df_new)).value_counts()
```

```python
wcss = []
for i in range(1,10) :
    KM = KMeans(n_clusters=i , random_state=78)
    KM.fit(df_new)
    wcss.append(KM.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
KM = KMeans(n_clusters=7 , random_state=78)
KM.fit(df_new)
KM_clusters = KM.predict(df_new)
```

```python
df_new.plot(x="Alcohol",y ="Color",c=KM.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans') 
```

```python
df_new.columns
```

```python
sns.scatterplot(data = df_new , x = "Dilution" , y = "Phenols" , hue=KM_clusters)
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_new, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show() 
```
