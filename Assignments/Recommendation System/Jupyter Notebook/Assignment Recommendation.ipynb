```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
df["genre"] = LE.fit_transform(df["genre"])
```

```python
df["type"] = LE.fit_transform(df["type"])
```

```python
def construct_data(data) :
    data_dict = {}
    row_names = ["genre","type","episodes","rating","members"]
    movie_names = list(data["name"])
    for movie in movie_names :
        if(movie not in data_dict.keys()) :
            data_dict[movie] = []
        else :
            continue
        for  i in row_names :
            my_data = float(data[data["name"] == movie][i].values[0])
            data_dict[movie].append(my_data)
    return data_dict      
data_dict = construct_data(df)
    
```

```python
new_df = pd.DataFrame(data_dict)
```

```python
new_df.info()
```

```python
def movie_recomandation(movie , data , no_of_movies = 5) :    
    movie_data = data[movie]
    similar_to_movie = new_df.corrwith(movie_data) 
    similar_df = pd.DataFrame(similar_to_movie, columns=['Correlation'])
    return list(similar_df[similar_df['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(no_of_movies).index)
```

```python
 movie_recomandation("vivi" , new_df , 10)
```

```python
 movie_recomandation("Robot King" , new_df , 10)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
sim = 1 - pairwise_distances(df[["genre","type","episodes","rating","members"]].values,metric='cosine')
```

```python
sim
```

```python
sim_df = pd.DataFrame(sim)
```

```python
sim_df.columns = df.name.values
```

```python
sim_df.index = df.name.values
```

```python
sim_df
```

```python
def recommend_cosine(movie , data , no_of_movies) :
    return list(data.idxmax(axis=1)[0:no_of_movies].values)
```

```python
recommend_cosine("Robot King" , sim_df , 10)
```

```python
#Collaborative Filtering
#Collaborative filtering is a method used by recommendation systems to make automatic predictions about a user's
#interests by collecting preferences or taste information from many users.
#The main assumption behind the collaborative filtering approach is that if a person A has the same opinion as a person B on an issue,
#A is more likely to have B's opinion on a different issue than that of a randomly chosen person. 


#User-Based Collaborative Filtering
#In user-based collaborative filtering, the system recommends 
#items by finding users similar to the target user and recommending items that those similar users have liked.
# 1)Finding Similar Users
# 2)Generating Recommendations


#Differences Between User-Based and Item-Based Collaborative Filtering
#Similarity Calculation:

#User-Based: Calculates similarity between users.
#Item-Based: Calculates similarity between items.

#Scalability:

#User-Based: Can be less scalable as the number of users increases.
#Item-Based: Generally more scalable, especially in scenarios with a large number of users but fewer items.
```
