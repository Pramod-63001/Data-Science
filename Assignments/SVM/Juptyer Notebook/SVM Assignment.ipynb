```python
X = df.drop("class" , axis=1)
y = df["class"]
```

```python
X = pd.get_dummies(X , columns=X.select_dtypes(["object"]).columns, drop_first=True ,dtype=np.int32 )
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
y = LE.fit_transform(y)
```

```python
# Since we have many features, let's reduce dimensions for visualization purposes using PCA
from sklearn.decomposition import PCA

no_of_comp = 3

pca = PCA(n_components=no_of_comp)
X_pca = pd.DataFrame(data = pca.fit_transform(X) , columns=[f"PCA_{i}" for i in range(no_of_comp)])
```

```python
from sklearn.model_selection import train_test_split
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
def error_matrix(X_train , X_test , y_train , y_test , model) :
    from sklearn.metrics import accuracy_score , f1_score , recall_score , precision_score
    y_test_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    print(f"Training Accuracy : {round(accuracy_score(y_train , y_train_pred),2)}")
    print(f"Testing Accuracy : {round(accuracy_score(y_test , y_test_pred),2)}")
    
    print(f"Training f1_score : {round(f1_score(y_train , y_train_pred),2)}")
    print(f"Testing f1_score : {round(f1_score(y_test , y_test_pred),2)}")
    
    print(f"Training recall_score : {round(recall_score(y_train , y_train_pred),2)}")
    print(f"Testing recall_score : {round(recall_score(y_test , y_test_pred),2)}")
    
    print(f"Training precision_score : {round(precision_score(y_train , y_train_pred),2)}")
    print(f"Testing precision_score : {round(precision_score(y_test , y_test_pred),2)}")
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

# Implement and train SVM classifier
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_pca, y_train_pca)
# Evaluate model performance on the testing data
error_matrix(X_train_pca, X_test_pca, y_train_pca, y_test_pca , svm_model)
```

```python
def conf_matrix(X , y_true , model) :
    from sklearn.metrics import confusion_matrix , ConfusionMatrixDisplay
    y_pred = model.predict(X)
    y_pred = [round(value) for value in y_pred]
    cm = confusion_matrix(y_true , y_pred)
    print("Confusion Metrix for Data")
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
```

```python
# Grid search for hyperparameter tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto']
}

grid_search = GridSearchCV(SVC(), param_grid, refit=True, verbose=2 , scoring="accuracy")
grid_search.fit(X_train_pca, y_train_pca)
```

```python
# Print the best parameters and best score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```

```python
# Compare SVM performance with different kernels
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    svm_model1 = SVC(kernel=kernel)
    svm_model1.fit(X_train_pca, y_train_pca)
    y_pred = svm_model1.predict(X_test_pca)
    print(f"Kernel: {kernel}")
    error_matrix(X_train_pca, X_test_pca, y_train_pca, y_test_pca , svm_model1)
    
#     print(classification_report(y_test, y_pred))
```

```python
#  Best Model
best_svm_model = SVC(C =  10, gamma =  'scale', kernel =  'rbf')
best_svm_model.fit(X_train_pca, y_train_pca)
# y_pred = berst_svm_model.predict(X_test_pca)
error_matrix(X_train_pca, X_test_pca, y_train_pca, y_test_pca , best_svm_model)
```

```python
conf_matrix(X_train_pca , y_train_pca , best_svm_model)
```
