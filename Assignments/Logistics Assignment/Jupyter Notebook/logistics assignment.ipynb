```python
df1 = pd.get_dummies(df1 , columns=["Sex", "Embarked"], drop_first=True ,dtype=np.int32 )
```

```python
df1.head()
```

```python
X = df1.drop("Survived" , axis = 1)
```

```python
y = df1["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
```

```python
for i in ["Age" , "Fare"] :
    X_train[i] = SS.fit_transform(X_train[[i]])
    X_test[i] = SS.transform(X_test[[i]])    
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LogR = LogisticRegression()
```

```python
LogR.fit(X_train , y_train)
```

```python
train_pred = LogR.predict(X_train)
```

```python
test_pred = LogR.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report , ConfusionMatrixDisplay , roc_curve
```

```python
print(classification_report(y_train , train_pred))
```

```python
# cm = confusion_matrix(y_train , train_pred)
print("Confusion Metrix for Training Data")
cm = confusion_matrix(y_train, train_pred, labels=LogR.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=LogR.classes_)
disp.plot()
plt.show()
```

```python
# cm = confusion_matrix(y_train , train_pred)
print("Confusion Metrix for testing Data")
cm = confusion_matrix(y_test, test_pred, labels=LogR.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=LogR.classes_)
disp.plot()
plt.show()
```

```python
print("Accuracy score for training Data : ", round(accuracy_score(y_train , train_pred),2))
```

```python
print("Accuracy score for testing Data : ", round(accuracy_score(y_test , test_pred),2))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
print("Roc Curve for training Data")
Logit_roc_score=roc_auc_score(y_train , train_pred)
fpr, tpr, thresholds = roc_curve(y_train,LogR.predict_proba(X_train)[:,1]) 
plt.plot(fpr, tpr, label='Logistic Regression (area=%0.2f)'% Logit_roc_score)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show() 
```

```python
print("Roc Curve for testing Data")
Logit_roc_score=roc_auc_score(y_train , train_pred)
fpr, tpr, thresholds = roc_curve(y_test,LogR.predict_proba(X_test)[:,1]) 
plt.plot(fpr, tpr, label='Logistic Regression (area=%0.2f)'% Logit_roc_score)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show() 
```
