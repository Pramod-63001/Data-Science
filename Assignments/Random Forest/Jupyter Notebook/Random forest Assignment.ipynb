```python
X = df.drop("Type" , axis = 1)
```

```python
y = df["Type"]
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
SS_X = scaler.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = X.columns
SS_X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=200, max_samples=0.8, max_features=0.8,random_state=42)
RF.fit(X_train, y_train)

y_pred_train = RF.predict(X_train)
y_pred_test = RF.predict(X_test)

from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print("RF - Training Accuracy:", train_accuracy)
print("RF - Testing Accuracy:", test_accuracy)
```

```python
from sklearn.ensemble import BaggingClassifier
bagging = BaggingClassifier(estimator=None, n_estimators=200, max_samples=0.8, max_features=0.8,random_state=42)
bagging.fit(X_train, y_train)

y_pred_train = bagging.predict(X_train)
y_pred_test = bagging.predict(X_test)

from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print("Bagging - Training Accuracy:", train_accuracy)
print("Bagging - Testing Accuracy:", test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report , ConfusionMatrixDisplay , roc_curve , f1_score
y_test_pred = RF.predict(X_test)
y_train_pred = RF.predict(X_train)

cm = confusion_matrix(y_train , y_train_pred)
print("Confusion Metrix for Training Data")
cm = confusion_matrix(y_train, y_train_pred, labels=RF.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=RF.classes_)
disp.plot()
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
boosting = AdaBoostClassifier(estimator=None, n_estimators=50,random_state=42)
boosting.fit(X_train, y_train)
```

```python
y_pred_train = boosting.predict(X_train)
y_pred_test = boosting.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)
```

```python
print("Boosting - Training Accuracy:", train_accuracy)
print("Boosting - Testing Accuracy:", test_accuracy)
```
