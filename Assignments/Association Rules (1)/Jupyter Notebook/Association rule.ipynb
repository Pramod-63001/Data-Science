```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings("ignore")
```

```python
# pip install mlxtend
```

```python
df = pd.read_excel("Online retail.xlsx" , header=None)
```

```python
col_names = df.iloc[: , 0].str.split(',', expand=True).iloc[0 , :]
```

```python
data_dict = {}
```

```python
for my_col in col_names :
    if(my_col not in data_dict.keys()) :
        data_dict[my_col] = []
    for data_row in df.iloc[: , 0] :
        if(my_col in data_row) :
            data_dict[my_col].append(1)
        else :
            data_dict[my_col].append(0)
```

```python
new_df = pd.DataFrame(data_dict)
```

```python
new_df
```

```python
frequently_itemsets = apriori(new_df , min_support = 0.005 , use_colnames=True)
```

```python
frequently_itemsets
```

```python
rules = association_rules(frequently_itemsets, metric="lift", min_threshold=0.8)
rules
rules.sort_values('lift',ascending = False).head(10)
```

```python
# An leverage value of 0 indicates independence. Range will be [-1 1]
# A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
lift_greater_1 = rules[rules.lift>1]
```

```python
lift_greater_1
```

```python
lift_greater_1["antecedents"]
```

```python
def func(data, *quant_name) :
    for quant in quant_name :
        if quant in data :
            return True
    return False

lift_greater_1[lift_greater_1["antecedents"].apply(func , args = ("shrimp","almonds"))]
```

```python
#1. What is Lift and Why is it Important in Association Rules?
#Lift is a measure used in association rule mining to evaluate the strength and importance of an association rule. It is defined as the ratio of the observed support to the expected support if the items were independent. Mathematically, for a rule 
#A-->B
#Lift(A→B)= Support(A∪B)/Support(A)×Support(B)

#2. What is Support and Confidence? How Do You Calculate Them?
#Support and Confidence are two key metrics used in association rule mining to measure the usefulness and accuracy of rules.
#Support: It measures how frequently an itemset appears in the dataset. For an itemset A, support is defined as:
#Support(A)= Number of transactions containing A/Total number of transactions
#Confidence: It measures how often items in B appear in transactions that contain A. For a rule A→B, confidence is defined as:
#Confidence(A→B)= Support(A∪B)/Support(A)

#3. What are Some Limitations or Challenges of Association Rule Mining?
#Limitations and Challenges:Scalability
#Redundant Rules
#Low Support Problem
#Actionability
#Handling of Rare Items
#Sparsity
```
