```python
# Bining
```

```python
df = df.drop("cc2_miles" , axis=1)
```

```python
df = df.drop("cc3_miles" , axis=1)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
k_bin = KBinsDiscretizer(random_state=58)
b_miles = k_bin.fit_transform(df[["Bonus_miles"]]).toarray()
b_miles = b_miles[:,1:]
```

```python
df = pd.concat([df , pd.DataFrame(b_miles , columns=[f'Bonus_miles_{i}' for i in range(b_miles.shape[1])] )] , axis = 1 )
```

```python
df = df.drop("Bonus_miles",axis = 1)
```

```python
df["Bonus_trans"].plot(kind = "hist")
```

```python
for cols in ["Bonus_trans","Flight_miles_12mo","Flight_trans_12"]:  
    binning = k_bin.fit_transform(df[[cols]]).toarray()
    binning = binning[:,1:] # To drop first calumn as one hot encoidng is being used
    df = pd.concat([df , pd.DataFrame(binning , columns=[f'{cols}_{i}' for i in range(binning.shape[1])] )] , axis = 1 )
```

```python
df = df.drop(["Bonus_trans","Flight_miles_12mo","Flight_trans_12"] , axis =1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MMS = MinMaxScaler()
```

```python
for cols in ["Balance","Days_since_enroll"] :
    df[cols] = MMS.fit_transform(df[[cols]])
```

```python
df.sample(5)
```

```python
from sklearn.cluster import KMeans
```

```python
KM = KMeans()
```

```python
KM.fit(df)
```

```python
pd.Series(KM.predict(df)).value_counts()
```

```python
wcss = []
for i in range(1,10) :
    KM = KMeans(n_clusters=i , random_state=78)
    KM.fit(df)
    wcss.append(KM.inertia_)
```

```python
plt.plot(range(1,10), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
df.plot(x="Balance",y ="Days_since_enroll",c=KM.labels_,kind="scatter",s=30 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans') 
```

```python
KM = KMeans(n_clusters= 7 , random_state=78)
KM.fit(df)
y_pred = KM.predict(df)
df["Y_KM"] = y_pred
```

```python
df
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df.drop("Y_KM" , axis =1), method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show() 
```

```python
raw_df = df.drop("Y_KM" , axis =1)
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=3, linkage='average',affinity = "euclidean").fit(raw_df) 

cluster_labels=pd.Series(h_complete.labels_)
df["Y_HA"] = cluster_labels
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(raw_df, method="single",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()    
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(raw_df, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show() 
```

```python
h_complete = AgglomerativeClustering(n_clusters=3, linkage='complete',affinity = "euclidean").fit(raw_df) 

cluster_labels=pd.Series(h_complete.labels_)
df["Y_HC"] = cluster_labels
```
